```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python

#types of values
#numerical
#categorical
#text
def column_analysis_numerical(column):
    count_nan = [tmp for tmp in column if np.isnan(tmp)]
    non_nan = [tmp for tmp in column if not np.isnan(tmp)]
    print('total', len(column))
    print('nan', len(count_nan))
    plt.hist(non_nan, bins=50)
```
